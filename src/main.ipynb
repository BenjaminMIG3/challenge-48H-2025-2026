```python
import pandas as pd 
from datetime import datetime
from time import strptime
```

```python
df = pd.read_csv('filtered_tweets_engie.csv', sep=';')
print(df.shape) # (576, 5)
df.head(10) 
```

```python
df['date'] = pd.to_datetime(df['created_at'])
df.head()
```

```python
df.drop(columns=['id', 'screen_name', 'created_at'],axis=1, inplace=True) #le inplace sert a du coup eviter de creer une variable "df", le inplace sert deja a mettre les valeurs dans le tableau 
df.head() 
```

```python
df.rename(columns={'name' : "Nom",'full_text' : 'Messages', 'date' : 'Date_de_publication'}, inplace=True )
df.head()
```

```python
# Supposons que vos donnÃ©es sont dans un DataFrame appelÃ© df et que la colonne de date s'appelle 'date'
df['Date_de_publication'] = pd.to_datetime(df['Date_de_publication'], utc = True)

# Nombre de tweets par jour
tweets_per_day = df.groupby(df['Date_de_publication'].dt.date).size()

# Nombre de tweets par mois
tweets_per_month = df.groupby(df['Date_de_publication'].dt.to_period('M')).size()

print("Tweets par jour:\n", tweets_per_day)
print("Tweets par mois:\n", tweets_per_month)
```
