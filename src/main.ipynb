```python
import pandas as pd 
from dotenv import load_dotenv
import os
import asyncio  # NÃ©cessaire pour exÃ©cuter les requÃªtes async
from mistralai import Mistral
import json
import time as t
```

```python
# Charger les variables d'environnement
load_dotenv()

# RÃ©cupÃ©rer la clÃ© API
api_key = os.getenv("MISTRAL_API_KEY")
print(f"ClÃ© API rÃ©cupÃ©rÃ©e : {api_key}")

# VÃ©rifier que la clÃ© API est bien dÃ©finie
if not api_key:
    raise ValueError("La clÃ© API MISTRAL_API_KEY n'est pas dÃ©finie dans l'environnement.")
```

```python
# Initialiser le nouveau client Mistral (asynchrone)
client = Mistral(api_key=api_key)
```

```python
def chat_bot(data: pd.DataFrame) -> pd.DataFrame:
    
    cpt = 0

    Sentiment = []
    Urgence = []
    Probleme = []
    Mot =[]
    inconfort =[]
    tipe = []
    Deception =[]
    Satisfaction =[]
    Frustration =[]

    for row in data.iterrows():
        cpt += 1
        t.sleep(2)
        
        response =  client.agents.complete(
            agent_id = "ag:753184c9:20250310:untitled-agent:dd78dfa6",  # Remplace par le bon modÃ¨le
            messages = [{"role": "user", "content": row[1]['Messages']}],
            response_format = {"type" :"json_object"}
        )

        # Transformation de la reponse en json
        resp = response.choices[0].message.content.strip()
        response_json = json.loads(resp)

        # Ajout des Ã©lÃ©ments de rÃ©ponse Ã  leur tableau respectif
        try:
            Sentiment.append(response_json['Sentiment'])
        except:
            Sentiment.append("")
        try:
            Urgence.append(response_json['Urgence'])
        except:
            Urgence.append("")
        try:
            Probleme.append(response_json['ProblÃ¨mes'])
        except:
            Probleme.append("")
        try:
            Mot.append(response_json['mot-critique'])
        except:
            Mot.append("")
        try:
            inconfort.append(response_json["Pourcentage d'inconfort"])
        except:
            inconfort.append("")
        try:
            tipe.append(response_json["Type de rÃ©clamation"])
        except:
            tipe.append("")
        try:
            Deception.append(response_json['DÃ©ception'])
        except:
            Deception.append("")
        try:
            Satisfaction.append(response_json['Satisfaction'])
        except:
            Satisfaction.append("")
        try:
            Frustration.append(response_json['Frustration'])
        except:
            Frustration.append("")

        # print(response_json["Sentiment"])
        # print(response_json["ProblÃ¨mes"])
        # print(response_json["Urgence"])
        # print(response_json["mot-critique"])
        # print(response_json["Pourcentage d\'inconfort"])
        # print(response_json["Type de rÃ©clamation"])
        # print(response_json["DÃ©ception"])
        # print(response_json["Satisfaction"])
        # print(response_json["Frustration"])
        print(f"{str(cpt)}/576 : OK")
    
    # Ajout des nouvelles colonnes dans le dataframe
    data['Sentiment'] = Sentiment
    data['Urgence'] = Urgence
    data['Probleme'] = Probleme
    data['mot-critique'] = Mot
    data['incomfort'] = inconfort
    data['type'] = tipe
    data['Deception'] = Deception
    data['Satisfaction'] = Satisfaction
    data['Frustration'] = Frustration

    return data
```

```python
df = pd.read_csv('filtered_tweets_engie.csv', sep=';')
print(df.shape) # (576, 5)
df.head(10) 
```

```python
df['date'] = pd.to_datetime(df['created_at'], utc=True)
df.head()
```

```python
df.drop(columns=['id', 'screen_name', 'created_at'],axis=1, inplace=True) #le inplace sert a du coup eviter de creer une variable "df", le inplace sert deja a mettre les valeurs dans le tableau 
df.head() 
```

```python
df.rename(columns={'name' : "Nom",'full_text' : 'Messages', 'date' : 'Date_de_publication'}, inplace=True )
df.head()
```

```python
# Nombre de tweets par jour
tweets_per_day = df.groupby(df['Date_de_publication'].dt.date).size()
print("Tweets par jour:\n", tweets_per_day)
```

```python
# Nombre de tweets par semaine
tweets_per_week = df.groupby(df['Date_de_publication'].dt.to_period('W')).size()
print("Tweets par semaine:\n", tweets_per_week)
```

```python
# Nombre de tweets par mois
tweets_per_month = df.groupby(df['Date_de_publication'].dt.to_period('M')).size()
print("Tweets par mois:\n", tweets_per_month)
```

```python
# FrÃ©quence des mentions des comptes Engie
engie_mentions = df['Messages'].str.contains('@Engie', case=False, na=False).sum()
print("FrÃ©quence des mentions des comptes Engie:", engie_mentions)
```

```python
# Liste des mots-clÃ©s critiques
keywords = ["dÃ©lai", "panne", "urgence", "scandale"]
keywords_regex = '|'.join(keywords)

# DÃ©tection des tweets contenant des mots-clÃ©s critiques
critical_tweets = df['Messages'].str.contains(keywords_regex, case=False, na=False).sum()
print("Nombre de tweets contenant des mots-clÃ©s critiques:", critical_tweets)
```

```python
df = chat_bot(df)
df.to_csv("Cleaned_file.csv", sep=";", encoding='utf-8', index=False, header=True)
print(df.shape)
df.head()
# Temps : 
```
